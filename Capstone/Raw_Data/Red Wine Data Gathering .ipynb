```python
import time
from selenium import webdriver
#opens a fake chrome browser
driver = webdriver.Chrome('/Users/divyasusarla/Downloads/chromedriver')  # Optional argument, if not specified will search path.
#starting link
driver.get("https://www.vivino.com/explore?e=eJzLLbI1VMvNzLM1UMtNrLA1NTBQS660LS1WKy-JjrU1BACbZwnB")
```

```python
card_filepath = '/Users/divyasusarla/Desktop/DSI-SF-2-divyasusarla/Capstone/cards_pickled.txt'
```

```python
#counts how many product 'cards' there are
cards = driver.find_elements(by='xpath', value="//div[@class='card card-lg']")
previous_save = 0
next_save = 1
while len(cards) < 10000:
    print len(cards)
    
    #the auto scroll
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    
    #check the total card count again
    cards = driver.find_elements(by='xpath', value="//div[@class='card card-lg']")
    
    # --- pickle / save every 500 / 1000
    # pickle = save any object, data frame, list, or dictionary
    if len(cards) > next_save:
        if next_save == 1:
            f = open(card_filepath, 'w')
        else:
            f = open(card_filepath, 'a+')
        links = [x.find_element_by_tag_name("a").get_attribute('href') for x in cards[previous_save:]]
        
        f.write('\n'.join(links)+'\n')
        f.close()
        previous_save = next_save
        next_save += 100
    
    time.sleep(3)
```
