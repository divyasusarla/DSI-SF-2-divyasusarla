```python
import pandas as pd
import os
```

```python
redwine = pd.DataFrame()

folder_path = '/Users/divyasusarla/Desktop/DSI-SF-2-divyasusarla/Capstone/redwinedetails/'
for x in os.listdir(folder_path):
    if 'csv' in x:
        localdf = pd.read_csv(folder_path + x)
        redwine = pd.concat([redwine, localdf], axis=0)
redwine.reset_index(inplace=True)
```
