```python
from scrapy.selector import Selector
import requests
from selenium import webdriver
import time
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
```

```python
whitewine_filepath = '/Users/divyasusarla/Desktop/DSI-SF-2-divyasusarla/Capstone/whitewine.txt'
```

```python
f = open(whitewine_filepath, 'r')
listoflinks = f.readlines()
f.close()
```

```python
ww_deduped = list(set(listoflinks))
```

```python
len(ww_deduped)
```

```python
driver = webdriver.Chrome('/Users/divyasusarla/Downloads/chromedriver')
i = 0
for path in ww_deduped:
    white_reviews_list = []
    i += 1

    driver.get(path)
    HTML = driver.page_source
    
    whitewine = Selector(text=HTML).xpath("//span[@class='text-block header-medium wine-info__title']/span[@class='semi wine-info__wine']/text()").extract()
    white_users = Selector(text=HTML).xpath("//a[@class='information__name']/text()").extract()
    white_reviews = Selector(text=HTML).xpath("//div[@class='review__content']/text()").extract()
    
    print whitewine
    
    for wine in whitewine:
        for users,reviews in zip(white_users, white_reviews):
            white_reviews_list.append([wine, users, reviews])
    
    df = pd.DataFrame(white_reviews_list)
    
    save_csv = '/Users/divyasusarla/Desktop/DSI-SF-2-divyasusarla/Capstone/whitewinereviews/whitereview_%d.csv' % i
    df.to_csv(save_csv, encoding='utf-8')
    
    time.sleep(2*np.random.rand())
driver.quit()
```
