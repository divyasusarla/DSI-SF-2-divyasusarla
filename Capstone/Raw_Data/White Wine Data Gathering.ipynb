```python
import time
from selenium import webdriver

driver = webdriver.Chrome('/Users/divyasusarla/Downloads/chromedriver')
#starting link
driver.get("https://www.vivino.com/explore?e=eJzLLbI1VMvNzLM1UMtNrLA1NTBQS660LS1WKy-JjrU1AgCbaAnC")
```

```python
whitewine_filepath = '/Users/divyasusarla/Desktop/DSI-SF-2-divyasusarla/Capstone/whitewine.txt'
```

```python
#counts how many product 'cards' there are
cards = driver.find_elements(by='xpath', value="//div[@class='card card-lg']")
previous_save = 0
next_save = 1
while len(cards) < 10000:
    print len(cards)
    
    #the auto scroll
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    
    #check the total card count again
    cards = driver.find_elements(by='xpath', value="//div[@class='card card-lg']")
    
    #save every 100
    if len(cards) > next_save:
        if next_save == 1:
            f = open(whitewine_filepath, 'w')
        else:
            f = open(whitewine_filepath, 'a+')
        links = [x.find_element_by_tag_name("a").get_attribute('href') for x in cards[previous_save:]]
        
        f.write('\n'.join(links)+'\n')
        f.close()
        previous_save = next_save
        next_save += 100
    
    time.sleep(3)
```

```python
f = open(whitewine_filepath, 'r')
whitewinelinks = f.readlines()
f.close()
```

```python
ww_deduped = list(set(whitewinelinks))
```

```python
len(ww_deduped)
```

```python
from scrapy.selector import Selector
import requests
```

```python
import numpy as np
import pandas as pd
```

```python
extract_list = []
i = 1
for path in ww_deduped:
    i +=1
    response = requests.get(path)
    HTML = response.text
    
    whitewine = Selector(text=HTML).xpath("//span[@class='text-block header-medium wine-info__title']/span[@class='semi wine-info__wine']/text()").extract()
    white_wineries = Selector(text=HTML).xpath("//span[@class='wine-info__winery']/a[@class='link-muted']/text()").extract()
    white_regions = Selector(text=HTML).xpath("//span[@class='text-block text-large wine-info__region']/span[@class='district']/a/text()").extract()
    white_countries = Selector(text=HTML).xpath("//span[@class='text-block text-large wine-info__region']/span[@class='country']/a/text()").extract()
    white_avg_wine_ratings = Selector(text=HTML).xpath("//span[@class='text-block header-smaller wine-info__average-info']/span[@class='average-info__rating']/text()").extract()
    white_price = Selector(text=HTML).xpath("//span[@class='text-block header-smaller wine-info__average-info']/span[@class='average-info__price']/text()").extract()
    
    
    extract_list.append([whitewine, white_wineries, white_regions, white_countries, white_avg_wine_ratings, white_price])
    if i % 100 == 0:
        print i, 'iteration'
        with open('whitewinedetails'+'000000'[:-len(str(i))]+str(i)+'.csv','wb') as f:
            df = pd.DataFrame(extract_list)
            df.to_csv(f)
            extract_list = []
    time.sleep(3*np.random.rand())
```
