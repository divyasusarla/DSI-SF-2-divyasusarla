```python
from scrapy.selector import Selector
import requests
from selenium import webdriver
import time
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
```

```python
card_filepath = '/Users/divyasusarla/Desktop/DSI-SF-2-divyasusarla/Capstone/cards_pickled.txt'
```

```python
f = open(card_filepath, 'r')
listoflinks = f.readlines()
f.close()
```

```python
rw_deduped = list(set(listoflinks))
```

```python
len(rw_deduped)
```

```python
driver = webdriver.Chrome('/Users/divyasusarla/Downloads/chromedriver')
i = 0
for path in rw_deduped:
    red_reviews_list = []
    i += 1
    # Swap out line 6 with the selenium chromedriver method and use the .page_content to assign the HTML variable
    driver.get(path)
    HTML = driver.page_source
    
    redwine = Selector(text=HTML).xpath("//span[@class='text-block header-medium wine-info__title']/span[@class='semi wine-info__wine']/text()").extract()
    red_users = Selector(text=HTML).xpath("//a[@class='information__name']/text()").extract()
    red_reviews = Selector(text=HTML).xpath("//div[@class='review__content']/text()").extract()
    
    print redwine
    
    for wine in redwine:
        for users,reviews in zip(red_users, red_reviews):
            red_reviews_list.append([wine, users, reviews])
    
    df = pd.DataFrame(red_reviews_list)
    
    save_csv = '/Users/divyasusarla/Desktop/DSI-SF-2-divyasusarla/Capstone/redwinereviews/redreview_%d.csv' % i
    df.to_csv(save_csv, encoding='utf-8')
    
    time.sleep(2*np.random.rand())
driver.quit()
```
